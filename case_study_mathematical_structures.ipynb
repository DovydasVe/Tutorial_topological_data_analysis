```python
from simplicial_complex import Simplex, SimplicialComplex
```

### The disk $B^2$

```python
disk_simplex = Simplex([0,1,2])
disk_sc = SimplicialComplex()
disk_sc.add_simplex(disk_simplex)

print("Allowed:", disk_sc.is_valid())
print(disk_sc.simplices)
print("Dimension:", disk_sc.dim())
```

### The circle $S^1$

```python
circle_simplex_1 = Simplex([0,1])
circle_simplex_2 = Simplex([0,2])
circle_simplex_3 = Simplex([1,2])
circle_list = [circle_simplex_1, circle_simplex_2, circle_simplex_3]

circle_sc = SimplicialComplex()
circle_sc.add_simplices(circle_list)

print("Allowed:", circle_sc.is_valid())
print(circle_sc.simplices)
print("Dimension:", circle_sc.dim())
```

### Tetrahedron boundary

```python
tetra_simplex_1 = Simplex([0,1,2])
tetra_simplex_2 = Simplex([0,1,3])
tetra_simplex_3 = Simplex([0,2,3])
tetra_simplex_4 = Simplex([1,2,3])
tetra_list = [tetra_simplex_1, tetra_simplex_2, tetra_simplex_3, tetra_simplex_4]

tetra_sc = SimplicialComplex()
tetra_sc.add_simplices(tetra_list)

print("Allowed:", tetra_sc.is_valid())
print(tetra_sc.simplices)
print("Dimension:", tetra_sc.dim())
```
